### Various CSV files combined to produce a clean US unemployment csv.
### CSV files used not included in github repo other than the final clean csv, which is used in the "features" notebook. 
### Therefore, code below won't run when attempted. 

```python
import pandas as pd
```

```python
def unemploy(stateID):
    data = stateID + '.csv'
    uemp = stateID + '_uemp.csv'
    df1 = pd.read_csv(data, header=None, squeeze=True)
    df_counties = pd.Series(df1.iloc[0,:])
    df_counties = df_counties.apply(lambda x: ' '.join(x.split(' ')[2:][:-1]))
    df_state = pd.Series(list(stateID  for x in range(len(df_counties))))
    df_cs = pd.DataFrame({'State':df_state,'County':df_counties})
    df_uemp = pd.read_csv(uemp)
    df_uemp['Series ID'] = df_uemp['Series ID'].apply(lambda x: x[-1])
    df_uemp = df_uemp[df_uemp['Series ID']=='3'].copy().reset_index()
    df_uemp = df_uemp.drop(columns=['index','Series ID'])
    df_data = pd.concat([df_cs, df_uemp], axis=1)
    return df_data    
```

```python
states = ['AL', 'AZ', 'AK', 'AR','CA', 'CO', 'CT', 'DE', 'FL','GA','HI', 'ID', 'IL','IN','IA',
         'KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC',
         'ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']
```

```python
states.sort()
```

```python
US_uemp = pd.DataFrame([])
for state in states:
#    print(state)
    US_uemp = pd.concat([US_uemp, unemploy(state)], axis=0)
```
